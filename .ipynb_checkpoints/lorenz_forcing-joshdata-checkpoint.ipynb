```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
```

```python
#Load Josh's data
data = np.load('./data/r0_0_L63.npy')[5000:50000+5000,:]
data_f1 = np.load('./data/r2_5_L63.npy')[5000:50000+5000,:]
data_f2 = np.load('./data/r4_0_L63.npy')[5000:50000+5000,:]
```

```python
x = data[:,0]
y = data[:,1]
z = data[:,2]

x_f1 = data_f1[:,0]
y_f1 = data_f1[:,1]
z_f1 = data_f1[:,2]

x_f2 = data_f2[:,0]
y_f2 = data_f2[:,1]
z_f2 = data_f2[:,2]


#Standardising data
xn = x/x.std()
yn = y/y.std()
zn = z/z.std()

xn_f1 = x_f1/x_f1.std()
yn_f1 = y_f1/y_f1.std()
zn_f1 = z_f1/z_f1.std()

xn_f2= x_f2/x_f2.std()
yn_f2= y_f2/y_f2.std()
zn_f2= z_f2/z_f2.std()


#Sticking standardised data back in a vector for future density computations
X = np.vstack([xn,yn, zn])
Xf1 = np.vstack([xn_f1,yn_f1, zn_f1])
Xf2 = np.vstack([xn_f2,yn_f2, zn_f2])
```

```python
# Visualize the strange attractor in 3D
elev = 10
azim = -40
xlims = (-3,3)
ylims = (-3,3)
zlims = (0,6)

N = 30000
s = 0.4
a = 0.2

fig = plt.figure(figsize=(10, 8))

#fig, axs = plt.subplots(2, 1, layout="constrained")

ax1 = fig.add_subplot(211, projection='3d')
ax1.scatter(xn[:N], yn[:N], zn[:N], c='k', s=s, alpha=a)  # Adjust s for point size
ax1.view_init(elev=elev, azim=azim)
ax1.set_xlim(xmin=xlims[0],xmax=xlims[1])
ax1.set_ylim(ymin=ylims[0],ymax=ylims[1])
ax1.set_zlim(zmin=zlims[0],zmax=zlims[1])
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Z')
#ax1.zaxis.set_tick_params(labelbottom=False)
ax1.set_title('(a) Lorenz `63 ($R = 49.5\%$)', y=0.9, fontsize=14)


ax2 = fig.add_subplot(212, projection='3d')
ax2.scatter(xn_f2[:N], yn_f2[:N], zn_f2[:N], c='k', s=s, alpha=a)  # Adjust s for point size
ax2.view_init(elev=elev, azim=azim)
ax2.set_xlim(xmin=xlims[0],xmax=xlims[1])
ax2.set_ylim(ymin=ylims[0],ymax=ylims[1])
ax2.set_zlim(zmin=zlims[0],zmax=zlims[1])
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('Z')
ax2.set_title('(b) Forced Lorenz `63 ($R = 62.9\%$)', y=0.9, fontsize=14)


#plt.figure(constrained_layout=True)
plt.tight_layout()
plt.savefig('/network/home/s/strommen/Python/Lorenz63/figures/lorenz_with_forcing_josh.jpg', format='jpg', dpi=300, bbox_inches='tight')
plt.show()
```

```python
from scipy import stats
kde = stats.gaussian_kde(X)
density = kde(X)

kde_f1 = stats.gaussian_kde(Xf1)
density_f1 = kde(Xf1)

kde_f2 = stats.gaussian_kde(Xf2)
density_f2 = kde(Xf2)
```

```python
density_diff = density_f2 - density
```

```python
fig = plt.figure(figsize=(10, 8))

#fig, axs = plt.subplots(2, 1, layout="constrained")

ax1 = fig.add_subplot(211, projection='3d')
ax1.scatter(xn, yn, zn, s=0.02, c=density, cmap='hot', alpha=0.5)  # Adjust s for point size
ax1.view_init(elev=elev, azim=azim)
ax1.set_xlim(xmin=xlims[0],xmax=xlims[1])
ax1.set_ylim(ymin=ylims[0],ymax=ylims[1])
ax1.set_zlim(zmin=zlims[0],zmax=zlims[1])
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Z')
#ax1.zaxis.set_tick_params(labelbottom=False)
ax1.set_title('(a) Lorenz `63', y=0.9, fontsize=16)


ax2 = fig.add_subplot(212, projection='3d')
#ax2.scatter(xn_f, yn_f, zn_f, c='k', s=0.01, alpha=0.3)
ax2.scatter(xn_f2, yn_f2, zn_f2, c=density_diff, cmap='RdBu_r', s=0.02, alpha=1) # Adjust s for point size
ax2.view_init(elev=elev, azim=azim)
ax2.set_xlim(xmin=xlims[0],xmax=xlims[1])
ax2.set_ylim(ymin=ylims[0],ymax=ylims[1])
ax2.set_zlim(zmin=zlims[0],zmax=zlims[1])
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('Z')
ax2.set_title('(b) Lorenz `63 with forcing term', y=0.9, fontsize=16)


#plt.figure(constrained_layout=True)
plt.tight_layout()
#plt.savefig('lorenz_with_forcing_vert-color.jpg', format='jpg', dpi=300, bbox_inches='tight')
plt.show()
```
